```python
#import all necessary libraries
import numpy as np
import pandas as pd
```

```python
#Load the data
df=pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/amazon.csv")
df.info()
# df
```

```python
#1) Handling with null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#2) Drop duplicate rows
df.drop_duplicates(inplace=True)
df.shape
```

Handling with each column in the table

```python
df["product_id"].unique()
# df.head()
```

```python
df["product_name"].unique()
```

```python
#Optional
#for product_name we need to do slicing as the product name is too big
df['product_name'] = df['product_name'].str.slice(0, 50)
df.head()
df['product_name'].unique()
```

```python
#renaming the about_product col to description
d={"about_product": "description"}
df.rename(columns=d,inplace=True)
df.head()
```

```python
#To remove the nan values
df['rating'] = df['rating'].replace('No rating available', np.nan)
df.head()
```

```python
#use inbuilt function title to captalise the each letter in front of the word
df['category'] = df['category'].str.title()
df.head()
```

```python
#remove several columns that are not necessary for the core analysis. I will drop the specified columns: user_id, user_name, review_id, review_title, review_content, img_link, product_link, and about_product.
columns_to_drop = ['user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link', 'description']
df.drop(columns=columns_to_drop, inplace=True)
df.head()
```

```python
#reseting the index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# df["product_id"].unique()
# df["product_name"].unique()
# df["category"].unique()
# df["discounted_price"].unique()
df["actual_price"].unique()
df["discount_percentage"].unique()
```

```python
#removing the 0% discount
df["discount_percentage"] = df["discount_percentage"].astype(str)
df["discount_percentage"] = df["discount_percentage"].str.strip("%")
df["discount_percentage"] = df["discount_percentage"].astype(float)
avg_discount = round(df["discount_percentage"].mean())
df["discount_percentage"] = df["discount_percentage"].replace(0.0, avg_discount)
df["discount_percentage"].unique()
```

```python
df["rating"].unique()
```

```python
# df['rating_count'].unique()
# df['rating_count'].isnull().sum()
# df.info()
#no null values so change data type
#change datatype to int
#to change this we need to remove the ',' in rating count
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = df['rating_count'].astype(int)
df["rating_count"].unique()

```
