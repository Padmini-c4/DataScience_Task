```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://github.com/Taj1920/Data_Analysis/raw/refs/heads/main/Youtube_analysis/YOUTUBE%20CHANNELS%20DATASET.csv")
df
```

#DATA CLEANING

```python
df.info()
df.isnull().sum()
```

```python
df["Ranking"].unique()
#as we have a nan value we use mean for rank
df["Ranking"].fillna(df["Ranking"].mean(),inplace=True)
df["Ranking"] = range(1, len(df) + 1)
df["Ranking"].unique()
```

```python
# df["Username"].unique()
df["Username"] = df["Username"].str.capitalize()
#we have a nan value in the username we use mode to do it
df["Username"].fillna(df["Username"].mode()[0],inplace=True)
df["Username"].unique()
```

```python
# df["Subscribers"].unique()
#here we do have nan value in subscriber
#first we need to do remove m by using strip
df["Subscribers"]=df["Subscribers"].str.strip('M')
#now we replace the nan value with the mode
df["Subscribers"].fillna(df["Subscribers"].mode()[0],inplace=True)
df["Subscribers"].unique()
```

```python
# df["Uploads"].unique()
#as we have comma in it we need to replace it with ""
# Convert to string first to ensure .str accessor can be used safely
df["Uploads"]=df["Uploads"].astype(str).str.replace(",","")
# df["Uploads"].unique()
#in the data we have nan
#now we need to convert the str to float datatype
df["Uploads"]=df["Uploads"].astype(float)

df["Uploads"] = df["Uploads"].fillna(df["Uploads"].mean())
df["Uploads"] = df["Uploads"].astype(int)

df["Uploads"].unique()
```

```python
df["Views"] = df["Views"].str.replace(",","")

mode = df["Views"].mode()
df["Views"] = df["Views"].fillna(0)


df["Views"] = df["Views"].astype(int)
df["Views"].unique()
```

```python
# df["Country"].unique()
#we have nan in the data so we use mode to replace it
df["Country"].fillna(df["Country"].mode()[0],inplace=True)
df["Country"].unique()
```

```python
df.info()
df.isnull().sum()
```

```python
#as we removed all the null values then we reset the index
df.reset_index(inplace=True)
df.drop("index",axis=1)
```

```python
data = df[["Subscribers", "Uploads", "Views"]].corr()
sns.heatmap(data,annot=True)
```

**1) Uploads do NOT determine subscribers**

*   Uploading more videos does NOT guarantee more subscribers
*   As we can see the numbering as 0.0031

**2)Subscribers highly relates to Views**


*   More subscribers can get easily more views for the videos that they post

**3)Uploads Vs Views have less relation**


**Insights :**


*  Uploading more videos does not always increase subscribers.
*   Channels with more subscribers usually get more total views.
*   Entertainment channels get good views.
*   News and TV channels usually upload many videos regularly.


**Recommendations**


*   Focus on good quality content, not just many videos.
*   More subscribers --> More views in the long run of a channel.
*   Try to make each video perform well, instead of only increasing the number of videos.
