Since we have null values we need to do Data Cleaning

2) Dataset Cleaning

```python
#Removing the unwanted columns
df.drop(["Unnamed: 0","url","address","phone","reviews_list","menu_item"],axis=1,inplace=True)
df
#we have removed these columns because these columns are not much suited to the data analysis
```

```python
#As we have two different columns with named of loctaion and listed_in(city),But the data is same
#when we consider which column has more no.of null values, we can see more no.of null values in location
#so lets drop the location column and rename the listed_in(city) with location
#Dropping the location
df.drop(["location"],axis=1,inplace=True)
df
#Renaming the listed_in(city)-->location
df.rename(columns={"listed_in(city)":"location"},inplace=True)
df
#Renaming name to Rest_name
df.rename(columns={"name":"rest_name"},inplace=True)
df
```

```python
#As we have two different columns with named of rest_type and listed_in(type),But the data is same,and listed_in(type) dont have the multiple values
#this makes the data engineer to retrive the data easily
#when we consider which column has more no.of null values, we can see more no.of null values in rest_type
#so lets drop the rest_type column and rename the listed_in(city) with rest_type
#Dropping the rest_type
df.drop(["rest_type"],axis=1,inplace=True)
df
#Renaming the listed_in(type)-->rest_type
df.rename(columns={"listed_in(type)":"rest_type"},inplace=True)
df
```

#Lets clean it with column by column

#rest_name

```python
#Now lets clean rest_name column
df["rest_name"].unique()
df.isnull().sum()
#rest_name col is perfect with zero null values
```

#online_order

```python
#online_order
df["online_order"].unique()
df.isnull().sum()
#online_order col is perfect with zero null values
```

#book_table

```python
#book_table
df["book_table"].unique()
df.isnull().sum()
#book_table col is perfect with zero null values
```

#	rate

```python
#rate
# df["rate"].unique()
#Removing the /5
df["rate"] = df["rate"].str.strip("/5")
# df["rate"].unique()
#Handling with null values
#i) replace NEW,-,nan with 0
df["rate"] = df["rate"].str.replace("NEW", "0")
df["rate"] = df["rate"].str.replace("-", "0")
df["rate"] = df["rate"].fillna('0')
# df["rate"].unique()
#ii) covert the str into float datatype
df["rate"] = df["rate"].astype(float)

#iii) replace 0 with mean
df["rate"]=df["rate"].replace(0,df["rate"].mean())
df["rate"].unique()
```

#votes

```python
#votes
df["votes"].unique()
df.isnull().sum()
#votes col is perfect with zero null values
```

#dish_liked

```python
#dish_liked
# df["dish_liked"].unique()
#here we have null values in dish_liked column(28078)
#using mode we need to replce that nan
df["dish_liked"]=df["dish_liked"].fillna(df["dish_liked"].mode()[0])
df.isnull().sum()
```

#cuisines

```python
#cuisines
# df["cuisines"].unique()
#here we have null values in cuisines column (45)
#using mode we need to replce that nan
df["cuisines"]=df["cuisines"].fillna(df["cuisines"].mode()[0])
df.isnull().sum()
```

#Cost_of_2

```python
#cost_of_2
#Renaming approx_cost(for two people) column to cost_of_2
df.rename(columns={"approx_cost(for two people)":"cost_of_2"},inplace=True)
# df["cost_of_2"].unique()
#here we have null values in cost_of_2 column (346)
#using mode we need to replce that nan
df["cost_of_2"]=df["cost_of_2"].fillna(df["cost_of_2"].mode()[0])
# df.isnull().sum()
#lets remove "," and replace it with ""
df["cost_of_2"]=df["cost_of_2"].str.replace(",","")
df["cost_of_2"] = df["cost_of_2"].astype(float)
df["cost_of_2"].unique()
```

#rest_type

```python
#rest_type
df["rest_type"].unique()
df.isnull().sum()
#rest_type col is perfect with zero null values
```

#Task 2: Univariate Analysis

**Objective:** Perform the following analysis with proper visualizations
1. Top 5 locations with highest number of restaurants


```python
#Fetching Top 5 locations with highest number of restaurants
#we can solve this by using values_count
top_loc=df["location"].value_counts().head()
# top_loc
#lets visualise

bar_chart=plt.bar(top_loc.index,top_loc.values,color="red",width=0.3)
plt.bar_label(bar_chart)
plt.xlabel("Location")
plt.ylabel("No. of Restaurants")
plt.title("Top 5 Locations with highest Number of Restaurants")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

*   Koramangala 7th Block	2938

2. Distribution of restaurant ratings

3. Percentage of restaurants offering online orders

```python
#By using pie chart we can perform Percentage of restaurants offering online orders
online_order_counts = df['online_order'].value_counts()
labels = online_order_counts.index
sizes = online_order_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightBlue'])
plt.title("Percentage of Restaurants Offering Online Orders")
plt.axis('equal')
plt.show()
```

Around 58.9% Resturants are offering online orders

and 41.1% are not giving the online orders
